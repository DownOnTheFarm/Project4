```python
import pandas as pd 
df = pd.read_csv("df.csv",header = 0)
df.head()
```

```python
y = df.pop('target')
```

```python
# Drop top 11 feature importance
dropped_df = df.drop(['mean_concave_points', 'worst_concave_points', 'worst_perimeter', 'worst_area', 'worst_radius', 'mean_concavity', 'worst_concavity', 'mean_perimeter', 'area_error', 'mean_area', 'mean_radius'],  axis=1)
dropped_df.head()
```

```python
X = dropped_df
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=78)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_scaler = scaler.fit(X_train)
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

```python
import tensorflow as tf
nn_model = tf.keras.models.Sequential()
nn_model.add(tf.keras.layers.Dense(units = 50, activation = 'relu' , input_dim = 19)) 
nn_model.add(tf.keras.layers.Dense(units = 50, activation = 'relu' , input_dim = 19)) 
nn_model.add(tf.keras.layers.Dense(units = 50, activation = 'relu' , input_dim = 19)) 
nn_model.add(tf.keras.layers.Dense(units = 1, activation = 'sigmoid'))
nn_model.summary()
```

```python
nn_model.compile(loss = 'binary_crossentropy' , optimizer = 'adam' , metrics = ['accuracy'])
```

```python
fit_model = nn_model.fit(X_train_scaled , y_train , epochs = 100)
```

```python
model_loss, model_accuracy = nn_model.evaluate(X_test_scaled, y_test, verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```
